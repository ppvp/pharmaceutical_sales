### R06 - Antihistamines for systemic use

```python
plt.figure(figsize=(15, 5))

df_antihist = df.groupby(['Year', 'Month'])['R06'].sum()
df_antihist.plot(kind='line', color='#D6B85A', marker='.')

ticks = [0, 12, 24, 36, 48, 60, 72]
labels = ['2014', '2015', '2016', '2017', '2018', '2019', '2020']

plt.title('[R06]', fontdict={'fontsize': 11})
plt.xlabel('Monthly Sales')
plt.ylabel('Sum of Sales (Volume)')
plt.xticks(ticks=ticks, labels=labels)

plt.show()
```

Sales of Antihistamines exhibit a repeating pattern with a steady increase in the volume of sales beginning with March, reaching its peak in June and a sharp decline in July and August. This indicates to us that the volume of sales of antihistamines is correlated with allergy season i.e. (Tree pollen in March/April, Grass pollen in June/July).

```python
plt.figure(figsize=(8, 8))

df_antihist_month = df.groupby(['Month'])['R06'].sum()
df_antihist_month.plot(kind='bar',color='#D6B85A', edgecolor='black', width=0.6, linewidth=2)

plt.title('[R06] Total Sales by Month [2014-2019]', fontdict={'fontsize': 11})
plt.xlabel('Month')
plt.xticks(rotation=0)
plt.ylabel('Total Sales (Volume)')

plt.show()
```

The months of April, May, and June, respectively, exhibit considerably greater sales numbers in contrast to the other months. Presumably due to the coincidence with allergy season (assuming Northern Hemisphere data origin).

# Monthly Sales Comparison

```python
plt.figure(figsize=(15, 9))

df_anti_infla_ts = df.groupby(['Year', 'Month'])['M01AB'].sum()
df_anti_infla_ts.plot(kind='line', color='#607c5c', marker='.')
df_anti_infla_ts_two = df.groupby(['Year', 'Month'])['M01AE'].sum()
df_anti_infla_ts_two.plot(kind='line', color='green', marker='.')
df_analgetics_ts = df.groupby(['Year', 'Month'])['N02BA'].sum()
df_analgetics_ts.plot(kind='line', color='#ff781f', marker='.')
df_analgetics_ts_two = df.groupby(['Year', 'Month'])['N02BE'].sum()
df_analgetics_ts_two.plot(kind='line', color='red', marker='.')
df_psych_ts = df.groupby(['Year', 'Month'])['N05B'].sum()
df_psych_ts.plot(kind='line', color='#281E5D', marker='.')
df_psych_ts_two = df.groupby(['Year', 'Month'])['N05C'].sum()
df_psych_ts_two.plot(kind='line', color='grey', marker='.')
df_air = df.groupby(['Year', 'Month'])['R03'].sum()
df_air.plot(kind='line', color='#609078', marker='.')
df_antihist = df.groupby(['Year', 'Month'])['R06'].sum()
df_antihist.plot(kind='line', color='#D6B85A', marker='.')

ticks = [0, 12, 24, 36, 48, 60, 72]
labels = ['2014', '2015', '2016', '2017', '2018', '2019', '2020']

plt.title('Sales Volume Comparison', fontdict={'fontsize': 16})
plt.xlabel('Monthly Sales')
plt.ylabel('Sum of Sales (Volume)')
plt.xticks(ticks=ticks, labels=labels)
plt.legend(bbox_to_anchor=(1.04, 1), loc="upper left")

plt.show()
```

# Daily Sales Volume


```python
weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

df['Weekday Name'] = pd.Categorical(df['Weekday Name'], categories=weekday_order, ordered=True)

df_day = df.groupby('Weekday Name')[df.columns[1:9]].sum()

df_day.plot(kind='bar', figsize=(17, 8))

plt.title('Sales Volume by Day [2014-2019]', fontdict={'fontsize': 16})
plt.xlabel('Day')
plt.xticks(rotation=0)
plt.ylabel('Total Sales (Volume)')
plt.legend(bbox_to_anchor=(1.04, 1), loc="upper left")


plt.show()
```
